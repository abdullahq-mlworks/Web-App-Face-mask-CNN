Project 14: Face Mask Detection (CNN)

```python
import tensorflow as trf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
```

```python
# Pehly hmary pas data arrays me hota ha lkn ab folders me images ha (png, jpg)

# rescale= -> ye asal me normalization hi ha lkn yaha hmary pas data folders me ha images ha to is lye 
# imagedatagenerator ke ander is trah likhy gy
# validation_split b ider hi likhy gy 
datagen= ImageDataGenerator(rescale= 1./255.0, validation_split= 0.2)
 
train_data= datagen.flow_from_directory("data",
                                        # resize the image into 128x128 q k hr inage different size ki ha dataset me
                                        target_size= (128, 128),
                                        # Ek time pe model 32 images process karega
                                        batch_size= 32,
                                        # binary 0,1  only 2 mask, without_mask
                                        class_mode= "binary",
                                        # Ye autometically 80% training data use karega
                                        subset= "training")
val_data= datagen.flow_from_directory("data",
                                        # resize the image into 128x128
                                        target_size= (128, 128),
                                        # Ek time pe model 32 images process karega
                                        batch_size= 32,
                                        # binary 0,1  only 2 mask, without_mask
                                        class_mode= "binary",
                                        # Ye autometically 80% training data use karega
                                        subset= "validation") # Ye training ke dauran checking hai
```

```python

model= Sequential()
# 1st layer
model.add(Conv2D(32, (3,3), activation="relu", input_shape= (128,128,3)))
model.add(MaxPooling2D(2,2))

# 2nd layer
model.add(Conv2D(64, (3,3), activation="relu"))
model.add(MaxPooling2D(2,2))

# 3rd layer
model.add(Conv2D(128, (3,3), activation="relu"))
model.add(MaxPooling2D(2,2))

model.add(Flatten())
model.add(Dense(128, activation="relu"))
model.add(Dense(64, activation="relu"))
model.add(Dense(1, activation="sigmoid")) # sigmoid -> for binary 0,1 nmask, without_mask
```

```python
model.compile(optimizer= "adam", loss= "binary_crossentropy", metrics= ["accuracy"])

model.fit(train_data, epochs= 5, validation_data= val_data)

loss, acc= model.evaluate(val_data)

print("Accuracy: ", acc)
print("Loss: ", loss)
```

```python
# .h5 = trained model ko file me store karne ka format
model.save("model.h5")
```
